```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_names = ['XGB','IFXGB','MLP','LOFMLP']
eval_metrics = ['Accuracy','Precision','Recall','F1','ROC','AUPRC','G-mean','MCC']
```

```python
scores = {'Accuracy': [0.9986, 0.9987, 0.9987, 0.9993],
          'Precision': [0.5396, 0.5641, 0.5538, 0.7613],
          'Recall': [0.8, 0.7764, 0.8470, 0.7882],
          'F1': [0.6445, 0.6534, 0.6697, 0.7745],
          'ROC': [0.9580, 0.9579, 0.9230, 0.8939],
          'AUPRC': [0.7152, 0.7072, 0.4693, 0.6004],
          'G-mean': [0.8939, 0.8807, 0.9198, 0.8876],
          'MCC': [0.6564, 0.6612, 0.6843, 0.7743]
}
```

```python
# Creating 8 individual plots for each metric
fig, axes = plt.subplots(4,2,figsize=(18,16)) 

fig.suptitle('Model Performance Comparison', fontsize=20, y=1.05)
axes = np.array(axes).flatten() # converts into 1D array to iterate through each subplot

for i, metrics in enumerate(eval_metrics):
    axes[i].plot(model_names, scores[metrics], marker='o', linestyle ='-', color='b')
    axes[i].set_title(metrics)
    axes[i].set_ylim(0.4, 1.05) # sets a common y-axis
    
    for j, k in enumerate(scores[metrics]):
        axes[i].text(j, k+0.02, "{:4f}".format(k), ha='center')

plt.tight_layout()
plt.show()
```
